# Data Visualization

### Customer Profile of the different Products

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Univariate Analysis

```python
#Distribution of Age using histogram
sns.histplot(x = "Age", data = fitness_df, kde = True)
plt.title("Age Distribution")
plt.xlabel("Customers' Age in Years")
plt.ylabel("Number of Customers")
plt.show()
```

Notes:
    
- The Age distribution of customers is right skewed
- Majority of the customers are of younger age, below 30 years old
- There are few customers that are of older age

```python
#Distribution of Product using countplot
sns.countplot(x = "Product", data = fitness_df)
plt.title("Product Distribution")
plt.xlabel("Treadmill Types (Model Number)")
plt.ylabel("Number of Customers")
plt.show()
```

- Over half of the total customers, which is 180, purchased the TM195 treadmill model, accounting for approximately 80 customers.

```python
#Distribution of Gender using countplot
sns.countplot(x = "Gender", data = fitness_df)
plt.title("Gender Distribution")
plt.xlabel("Gender Types")
plt.ylabel("Number of Customers")
plt.show()
```

- Of the 180 total customers, over 100 are male, representing the majority of the customer base.

```python
#Distribution of Marital Status using countplot
sns.countplot(x = "MaritalStatus", data = fitness_df)
plt.title("Plot of Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Number of Customers")
plt.show()
```

- The majority of customers, more than 100 people, have listed their marital status as "Partnered," while fewer than 80 people are listed as "Single."

```python
#Distribution of Income using boxplot
sns.boxplot(x = "Income", data = fitness_df)
plt.title("Income Distribution")
plt.xlabel("Income of Customers in GBP")
plt.ylabel("Number of Customers")
plt.show()
```

- On the average, the income of customers is about GBP50,000

```python
#Distribution of Usage using boxplot
sns.boxplot(x = "Usage", data = fitness_df)
plt.title("Usage Distribution")
plt.xlabel("Average Usage Per Week")
plt.ylabel("Number of Customers")
plt.show()
```

- On average, customers are expected to use the treadmill approximately 3 to 4 times per week.

#### Multivariate Analysis

- This plot shows that there is some relationship between the Age of customers and their income. Customers with higher age earn more.

```python
#Relationship between Age and Product Type
sns.boxplot(x = "Age", y = "Product", data = fitness_df)
plt.title("Plot of Age vs Product Type")
plt.xlabel("Age of Customers in Years")
plt.ylabel("Type of Treadmill (Model No)")
plt.show()
```

- Customers who purchase the TM798 treadmill model tend to be slightly older on average compared to those who purchase the TM498 and TM195 models.

```python
#Relationship between Income and Product Type
sns.boxplot(x = "Income", y = "Product", data = fitness_df)
plt.title("Plot of Income vs Product Type")
plt.xlabel("Income of Customers in GBP")
plt.ylabel("Type of Treadmill (Model No)")
plt.show()
```

- Customers who purchase the TM798 treadmill model have a significantly higher average income compared to those who purchase other models, with TM195 customers having the lowest average income.

```python
#Relationship between Usage and Product Type
sns.catplot(x = "Usage", y = "Product", data = fitness_df, kind ="box")
plt.title("Plot of Usage vs Product Type")
plt.xlabel("Average Usage Per Week")
plt.ylabel("Type of Treadmill (Model No)")
plt.show()
```

- The TM498 treadmill model is expected to have the lowest usage, while the TM798 model is expected to have the highest usage, slightly exceeding the average usage of the TM195 model.

```python
#Relationship between Age and Income
sns.scatterplot(x = "Age", y = "Income", data = fitness_df)
plt.title("Plot of Age vs Income")
plt.xlabel("Age of Customers in Years")
plt.ylabel("Income of Customers in GBP")
plt.show()
```

- This plot shows that there is some relationship between the Age of customers and their income. Customers with higher age earn more.

```python
#Relationship between Gender and Product Type
sns.scatterplot(y = "Gender", x = "Product", data = fitness_df)
plt.title("Plot of Gender vs Product Type")
plt.ylabel("Gender of Customers")
plt.xlabel("Type of Treadmill (Model No)")
plt.show()
```

- There is no correlation between the gender of customers and the model of the treadmill they purchase.

```python
#Relationship between Marital Status and Product Type
sns.scatterplot(y = "MaritalStatus", x = "Product", data = fitness_df)
plt.title("Plot of Marital Status vs Product Type")
plt.ylabel("Marital Status of Customers")
plt.xlabel("Type of Treadmill (Model No)")
plt.show()
```

- There is no correlation between the marital status of the customer and the specific model of treadmill purchased.

```python
#Relationship between Miles and Fitness
sns.scatterplot(y = "Miles", x = "Fitness", data = fitness_df)
plt.title("Plot of Fitness vs Miles")
plt.ylabel("Expected Miles")
plt.xlabel("Fitness of Customers")
plt.show()
```

- There is a positive correlation between a customer's fitness level and the expected distance they will run on the treadmill. Customers who are deemed very unfit are expected to run the shortest distance, while those who are very fit are expected to run the longest distance.

```python
#Relationship between Age and Fitness
sns.scatterplot(x = "Age", y = "Fitness", data = fitness_df)
plt.title("Plot of Age vs Fitness")
plt.xlabel("Age of Customers in Years")
plt.ylabel("Fitness of Customers")
plt.show()
```

```python
There is no correlation between the age and fitness of the customer. Even some older people are very fit
```

### Conclusion and Recommendation

After analyzing the data of 180 customers who purchased treadmill products from Cardio Good Fitness, we have arrived at the following conclusions and recommendations:

Findings:
        
- The TM195 model is the most popular treadmill model among customers.
- The majority of customers are males who are partnered.
- Most of the customers are younger than 30 years old, with few older customers.
- Over 50% of the total customers bought the TM195 model, which is approximately 80 customers.
- Customers who purchase the TM798 treadmill model tend to be slightly older than those who buy TM498 and TM195 models.
- Customers who buy the TM798 model have significantly higher average income compared to those who buy other models, with TM195 customers having the lowest average income.
- Of the 180 total customers, over 100 are males.
- There is a positive correlation between a customer's fitness level and the expected distance they will run on the treadmill, where very fit customers run the longest distance and very unfit customers run the shortest distance.

    
Recommendations:
        
- Production of TM195 model should be increased as it is the most purchased treadmill model.
- Cardio Good Fitness should target younger customers, especially those below 30 years old.
- Focus on male customers and partnered customers.
- High-income earners may not necessarily be the target market as customers who purchase TM195 model have below-average income of approximately GBP53,719.58.
